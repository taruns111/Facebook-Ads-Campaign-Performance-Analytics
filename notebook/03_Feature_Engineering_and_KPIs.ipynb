```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data\\cleaned_facebook_ads.csv')
```

```python
df.head()
```

# Click Through Rate (CTR)
CTR = (Clicks / Impressions) * 100

```python
df['CTR'] = (df['clicks'] / df['impressions']) * 100
```

# Click Per Click (CPC)
CPC = Spent / Clicks

```python
df['CPC'] = df['spent'] / df['clicks']
```

# Cost Per Mille (CPM)
Cost per 1000 impression

CPM = (Spent / Impression) * 1000

```python
df['CPM'] = (df['spent'] / df['impressions']) * 1000
```

# Conversion Rate
Conversion Rate = (Total_Conversion / clicks) * 100

```python
df['Conversion_rate'] = (
    df['Total_Conversion'] / df['clicks']
) * 100
```

# Approved conversion rate
Approved conversion rate = Approved_conversion / Total_conversion * 100

```python
df['Approved_Conversion_Rate'] = (
    df['Approved_Conversion'] /
    df['Total_Conversion']
) * 100
```

# Cost per conversion
CPA = Spent / Total_Conversion

```python
df['CPA'] = df['spent'] / df['Total_Conversion']
```

# Cost Per Approved Conversion
Spent / Approved_Conversion

```python
df['Cost_Per_Approved'] = (
    df['spent'] /
    df['Approved_Conversion']
)
```

# Conversion per 1000 Impressions


```python
df['Conversions_per_1000'] = (
    df['Total_Conversion'] /
    df['impressions']
) * 1000
```
